Согласно описанию дата сета, успеваемость - это три последних столбца: G1, G2, G3.
    
These grades are related with the course subject, Math or Portuguese:

G1 - first period grade (numeric: from 0 to 20);

G2 - second period grade (numeric: from 0 to 20);

G3 - final grade (numeric: from 0 to 20, output target).

Если я правильно понял, то G1 и G2 - это НЕ оценки по предметам (математика / португальский), а оценки по какому-то (?) предмету за полугодия (first / second period).
G3 - финальная оценка.
Поскольку в датасете нет других временных периодов для сопоставления G1 и G2, считаю, их можно исключить и использовать для оценки успеваемости только столбец G3.

Для определения уровня потребления алкоголя возьмём по каждому ученику среднее по столбцам Dalc и Walc. Корректнее было бы не просто посчитать среднее, а также учитывать удельный вес будних и выходных дней. Вероятно, удельный вес Workday (Dalc) больше, чем Weekend (Walc), но соотношение "На 5 рабочих дней приходится 2 дня уикенда" тут будет некорректным, т.к. тогда не учитываются праздники, каникулы и т.п. Поэтому просто рассчитаем среднее. Добавим столбец alco_mean.

```python
alcoweek = ['Dalc', 'Walc']
df2['alco_mean'] = df2[alcoweek].mean(axis=1)
df2
```

Сгруппируем сначала уровень потребления алкоголя по оценкам. Попробуем сделать какой-то вывод просто по списку значений - может быть зависимость сразу будет заметна.

```python
df2.groupby('G3')['alco_mean'].mean()
```

Проследить какую-то зависимость значений не удаётся. Сконвертируем полученный DataFrameGroupBy в DataFrame и попробуем сделать какой-то вывод с помощью визуализации. Используем столбчатые диаграммы (bar).

```python
alco_mean_by_g3 = df2.groupby('G3')['alco_mean'].mean().to_frame(name='alco_mean').reset_index()
alco_mean_by_g3.plot.bar(x='G3', y=['alco_mean'])
```

Более глубокий анализ можно провести, используя ящик с усами (boxplot).

```python
df2.boxplot(column='alco_mean', by='G3', figsize=(8, 8))
```

Боксплот в данном случае более наглядно показывает динамику, т.к. помимо медианы зависимость можно отслежить по верхнему краю ящика. Отдельно отмеченные точки (Outliers) являются выбросами значений. Есть небольшой нисходящий тренд значений потребления алкоголя при улучшении успеваемости, но в целом опеределить какую-то зависимость невозможно.
